# **Longest Substring Without Repeating Characters**

Input: s = "abcabcbb"

Output: 3

Explanation: The answer is "abc", with length 3.


```python
def lengthOfLongestSubstring(s):
    seen = {}
    left = 0
    max_len = 0

    for right, char in enumerate(s):
        if char in seen and seen[char] >= left:
            left = seen[char] + 1
        seen[char] = right
        max_len = max(max_len, right - left + 1)

    return max_len
```

```python
s = "abcabcbb"
print(lengthOfLongestSubstring(s))  # 3
```

# **String to Integer (atoi)**
Input: s = "42"
Output: 42

Input: s = "   -42"
Output: -42

Input: s = "4193 with words"
Output: 4193


```python
def myAtoi(s):
    s = s.lstrip()
    if not s:
        return 0
    sign = 1
    if s[0] in "+-":
        if s[0] == '-':
            sign = -1
        s = s[1:]
    num = 0
    for char in s:
        if char.isdigit():
            num = num * 10 + int(char)
        else:
            break
    num *= sign
    INT_MAX = 2**31 - 1
    INT_MIN = -2**31
    return max(min(num, INT_MAX), INT_MIN)
```

```python
print(myAtoi("42"))           # 42
print(myAtoi("   -42"))       # -42
print(myAtoi("4193 with words"))  # 4193
```

# **Implement strStr() (Substring Search)**
Input: haystack = "hello", needle = "ll"

Output: 2


```python
def strStr(haystack, needle):
    if not needle:
        return 0
    for i in range(len(haystack) - len(needle) + 1):
        if haystack[i:i+len(needle)] == needle:
            return i
    return -1
```

```python
haystack = "hello"
needle = "ll"
print(strStr(haystack, needle))  # 2
```

# **Group Anagrams**
Input: ["eat","tea","tan","ate",
"nat","bat"]

Output: [["eat","tea","ate"],["tan","nat"],["bat"]]


```python
from collections import defaultdict

def groupAnagrams(strs):
    anagrams = defaultdict(list)
    for s in strs:
        key = tuple(sorted(s))
        anagrams[key].append(s)
    return list(anagrams.values())
```

```python
strs = ["eat","tea","tan","ate","nat","bat"]
print(groupAnagrams(strs))
```

```python
def are_anagrams_key(str1, str2):

  str1, str2 = str1.replace(" ",""), str2.replace(" ","")
  if len(str1) != len(str2):
    return False

  dict_str_1 = {}
  dict_str_2 = {}

  for char in str1.lower():
    if char not in dict_str_1:
      dict_str_1[char] = 1
    else:
      dict_str_1[char] += 1

  for char in str2:
    if char not in dict_str_2:
      dict_str_2[char] = 1
    else:
      dict_str_2[char] += 1

  return dict_str_1 == dict_str_2
```

```python
def subsets(set):
  result = []
  total_num = len(set)

  def backtract(i, path):
    if i == total_num:
      result.append(path[:])
      return

    path.append(set[i])
    backtract(i+1, path)
    path.pop()

    backtract(i+1, path)

  backtract(0, [])
  return result
```

# **Set With Target**

```python
def target_subsets(set, target):
  result = []
  total_num = len(set)

  def backtract(i, sum, path):
    if i == total_num:
      if sum == target:
        result.append(path[:])
      return

    path.append(set[i])
    backtract(i+1, sum + set[i], path)
    path.pop()

    backtract(i+1, sum, path)

  backtract(0, 0, [])
  return result
```

```python
set_ = [1, 2, 3, 4, 5, 7]
target_subsets(set_, 10)
```

# **Binary String With Specific Lenght & Start With**

```python
def binary_strings_start_with_1(n):
    result = []

    def backtrack(path):
        if len(path) == n:
            result.append("".join(path))
            return

        path.append('0')
        backtrack(path)
        path.pop()

        path.append('1')
        backtrack(path)
        path.pop()

    backtrack(['1'])
    return result
```

# **2Sum Index**

```python
def two_sum_values(nums, target):
    seen = set()
    pairs = []
    for num in nums:
        compliment = target - num
        if compliment in seen:
            pairs.append([compliment, num])
        seen.add(num)
    return pairs
```

# **3Sum**
Input: nums = [-1,0,1,2,-1,-4]

Output: [[-1,-1,2],[-1,0,1]]


```python
def threeSum(nums):
    nums.sort()
    res = []
    n = len(nums)

    for i in range(n):
        if i > 0 and nums[i] == nums[i-1]:
            continue  # skip duplicates
        left, right = i+1, n-1
        while left < right:
            total = nums[i] + nums[left] + nums[right]
            if total < 0:
                left += 1
            elif total > 0:
                right -= 1
            else:
                res.append([nums[i], nums[left], nums[right]])
                left += 1
                right -= 1
                while left < right and nums[left] == nums[left-1]:
                    left += 1
                while left < right and nums[right] == nums[right+1]:
                    right -= 1
    return res
```

```python
nums = [-1,0,1,2,-1,-4]
print(threeSum(nums))  # [[-1,-1,2],[-1,0,1]]
```

```python
3Sum Closest
Input: nums = [-1,2,1,-4], target = 1
Output: 2
Explanation: -1 + 2 + 1 = 2
```

```python
def threeSumClosest(nums, target):
    nums.sort()
    closest = float('inf')

    for i in range(len(nums)-2):
        left, right = i+1, len(nums)-1
        while left < right:
            total = nums[i] + nums[left] + nums[right]
            if abs(total - target) < abs(closest - target):
                closest = total
            if total < target:
                left += 1
            elif total > target:
                right -= 1
            else:
                return total
    return closest
```

```python
nums = [-1,2,1,-4]
target = 1
print(threeSumClosest(nums, target))  # 2
```

Input: nums = [1,2,3,4]
Output: [24,12,8,6]


```python
def productExceptSelf(nums):
    n = len(nums)
    answer = [1]*n

    # Left products
    left = 1
    for i in range(n):
        answer[i] = left
        left *= nums[i]

    # Right products
    right = 1
    for i in range(n-1, -1, -1):
        answer[i] *= right
        right *= nums[i]

    return answer
```

```python
nums = [1,2,3,4]
print(productExceptSelf(nums))  # [24,12,8,6]
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
node15 = TreeNode(15)
node7 = TreeNode(7)
node9 = TreeNode(9)

node20 = TreeNode(20, node15, node7)

root = TreeNode(3, node9, node20)
```

```python
def maxDepth(root):
    if not root:
        return 0

    left = maxDepth(root.left)
    right = maxDepth(root.right)

    return 1 + max(left, right)  # âœ… This is correct
```

```python
from collections import deque

def maxDepthBFS(root):
    if not root:
        return 0

    queue = deque([root])
    depth = 0

    while queue:
        level_size = len(queue)  # Number of nodes at current level
        for _ in range(level_size):
            node = queue.popleft()
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        depth += 1  # Finished one level

    return depth
```

# **If Tree Is Symmetric**

        1     
       / \
      2   2         
     / \ / \
    3  4 4  3

        and

        1
       / \
      2   2
       \    \
        3    3



```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
root_symmetric = TreeNode(1,
                          TreeNode(2, TreeNode(3), TreeNode(4)),
                          TreeNode(2, TreeNode(4), TreeNode(3)))
```

```python
root_nonsymmetric = TreeNode(1,
                             TreeNode(2, None, TreeNode(3)),
                             TreeNode(2, None, TreeNode(3)))
```

```python
def isSymmetric(root):
    if not root:
        return True

    def isMirror(t1, t2):
        if not t1 and not t2:
            return True
        if not t1 or not t2:
            return False
        return (t1.val == t2.val) and \
               isMirror(t1.left, t2.right) and \
               isMirror(t1.right, t2.left)

    return isMirror(root.left, root.right)
```

```python
from collections import deque

def isSymmetricBFS(root):
    if not root:
        return True

    queue = deque()
    queue.append((root.left, root.right))

    while queue:
        t1, t2 = queue.popleft()

        if not t1 and not t2:
            continue
        if not t1 or not t2:
            return False
        if t1.val != t2.val:
            return False

        queue.append((t1.left, t2.right))
        queue.append((t1.right, t2.left))

    return True
```

```python
print(isSymmetric(root_symmetric))
print(isSymmetric(root_nonsymmetric))
```

```python
print(isSymmetricBFS(root_symmetric))      # Output: True
print(isSymmetricBFS(root_nonsymmetric))   # Output: False
```

# **Inverted Binary Tree**

        4
       / \
      2   7
     / \ / \
    1  3 6  9

        To
        
        4
       / \
      7   2
     / \ / \
    9  6 3  1


```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
root = TreeNode(4,
                TreeNode(2, TreeNode(1), TreeNode(3)),
                TreeNode(7, TreeNode(6), TreeNode(9)))
```

```python
def invertTree(root):
    if not root:
        return None

    root.left, root.right = root.right, root.left

    invertTree(root.left)
    invertTree(root.right)

    return root
```

```python
from collections import deque

def invertTreeBFS(root):
    if not root:
        return None

    queue = deque([root])
    while queue:
        node = queue.popleft()
        node.left, node.right = node.right, node.left  # Swap children

        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)

    return root
```

```python
def printTree(root):
    # Simple BFS print for demonstration
    from collections import deque
    queue = deque([root])
    while queue:
        node = queue.popleft()
        if node:
            print(node.val, end=' ')
            queue.append(node.left)
            queue.append(node.right)

print("Original Tree (BFS):")
printTree(root)

inverted_root = invertTree(root)
print("\nInverted Tree (BFS):")
printTree(inverted_root)
```

# **Binary Tree Level Order Traversal**

        3
       / \
      9   20
          / \
         15  7

Output ðŸ‡°

  [3],

  [15, 7]

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
```

```python
root = TreeNode(3,
                TreeNode(9),
                TreeNode(20, TreeNode(15), TreeNode(7)))
```

```python
def levelOrderDFS(root):
    result = []

    def dfs(node, depth):
        if not node:
            return
        # If we reach a new depth, add a new list
        if len(result) <= depth:
            result.append([])
        result[depth].append(node.val)
        # Recurse for children
        dfs(node.left, depth + 1)
        dfs(node.right, depth + 1)

    dfs(root, 0)
    return result
```

```python
from collections import deque

def levelOrderBFS(root):
    if not root:
        return []

    result = []
    queue = deque([root])

    while queue:
        level_size = len(queue)
        level_nodes = []
        for _ in range(level_size):
            node = queue.popleft()
            level_nodes.append(node.val)
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        result.append(level_nodes)

    return result
```

# **Reverse Linked List**

Input: 1 -> 2 -> 3 -> 4 -> 5 -> None

Output: 5 -> 4 -> 3 -> 2 -> 1 -> None


```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverseLinkedList(head):
    prev = None
    current = head
    while current:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
    return prev
```

```python
head = ListNode(1, ListNode(2, ListNode(3)))
new_head = reverseLinkedList(head)

current = new_head
while current:
    print(current.val, end=" -> ")
    current = current.next
print("None")
```

# **Detect A Cycle**

Input: 3 -> 2 -> 0 -> -4 (cycle connects -4 -> 2)

Output: True


```python
def hasCycle(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            return True
    return False
```

```python
head = ListNode(3, ListNode(2, ListNode(0, ListNode(-4))))
head.next.next.next.next = head.next  # create cycle
print(hasCycle(head))  # Output: True
```

# **Merge Two Sorted Lists**
Input: 1->2->4, 1->3->4

Output: 1->1->2->3->4->4


```python
def mergeTwoLists(l1, l2):
    dummy = ListNode(0)
    tail = dummy

    while l1 and l2:
        if l1.val < l2.val:
            tail.next = l1
            l1 = l1.next
        else:
            tail.next = l2
            l2 = l2.next
        tail = tail.next

    tail.next = l1 or l2
    return dummy.next
```

# **Sort a Linked List**
Input: 4 -> 2 -> 1 -> 3 -> None

Output: 1 -> 2 -> 3 -> 4 -> None


```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

# Function to merge two sorted linked lists
def merge(l1, l2):
    dummy = ListNode(0)
    tail = dummy
    while l1 and l2:
        if l1.val < l2.val:
            tail.next = l1
            l1 = l1.next
        else:
            tail.next = l2
            l2 = l2.next
        tail = tail.next
    tail.next = l1 or l2
    return dummy.next

# Function to find the middle of the linked list
def getMiddle(head):
    slow, fast = head, head.next
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    return slow

# Main function to sort linked list
def sortLinkedList(head):
    if not head or not head.next:
        return head

    mid = getMiddle(head)
    right_head = mid.next
    mid.next = None

    left_sorted = sortLinkedList(head)
    right_sorted = sortLinkedList(right_head)

    return merge(left_sorted, right_sorted)
```

```python
# Create unsorted linked list: 4 -> 2 -> 1 -> 3
head = ListNode(4, ListNode(2, ListNode(1, ListNode(3))))

# Sort the list
sorted_head = sortLinkedList(head)

# Print sorted linked list
current = sorted_head
while current:
    print(current.val, end=" -> ")
    current = current.next
print("None")
```

# **Remove N-th Node from End**
Input: 1->2->3->4->5, n = 2

Output: 1->2->3->5

```python
def removeNthFromEnd(head, n):
    dummy = ListNode(0, head)
    first = second = dummy

    # Move first n+1 steps ahead
    for _ in range(n + 1):
        first = first.next

    # Move both pointers
    while first:
        first = first.next
        second = second.next

    # Remove the n-th node
    second.next = second.next.next
    return dummy.next
```

```python
head = ListNode(1, ListNode(2, ListNode(3, ListNode(4, ListNode(5)))))
new_head = removeNthFromEnd(head, 2)

current = new_head
while current:
    print(current.val, end=" -> ")
    current = current.next
print("None")
```

# **Intersection of Two Linked Lists**

Input: A = 1->2->3->4->5, B = 6->4->5 (intersection at node with value 4)

Output: 4


```python
def getIntersectionNode(headA, headB):
    if not headA or not headB:
        return None

    a, b = headA, headB
    while a != b:
        a = a.next if a else headB
        b = b.next if b else headA
    return a
```

```python
common = ListNode(4, ListNode(5))
headA = ListNode(1, ListNode(2, ListNode(3, common)))
headB = ListNode(6, common)

intersection = getIntersectionNode(headA, headB)
print(intersection.val if intersection else "No intersection")
```

# **Add Two Numbers (as Linked Lists)**
Input: list1 = 2->4->3, list2 = 5->6->4

Output: 7->0->8  (342 + 465 = 807)


```python
def addTwoNumbers(l1, l2):
    dummy = ListNode(0)
    curr = dummy
    carry = 0

    while l1 or l2 or carry:
        x = l1.val if l1 else 0
        y = l2.val if l2 else 0
        total = x + y + carry
        carry = total // 10
        curr.next = ListNode(total % 10)
        curr = curr.next
        if l1: l1 = l1.next
        if l2: l2 = l2.next

    return dummy.next
```

# **Copy List With Random Pointer**

Input: Node1(val=7) -> Node2(val=13) -> Node3(val=11)

Random pointers: Node1.random=None, Node2.random=Node1, Node3.random=Node1

Output: Deep copy of the list with same structure and random pointers


```python
class RandomListNode:
    def __init__(self, val=0, next=None, random=None):
        self.val = val
        self.next = next
        self.random = random

def copyRandomList(head):
    if not head:
        return None

    # Step 1: Copy nodes and insert them next to original nodes
    curr = head
    while curr:
        new_node = RandomListNode(curr.val, curr.next)
        curr.next = new_node
        curr = new_node.next

    # Step 2: Copy random pointers
    curr = head
    while curr:
        if curr.random:
            curr.next.random = curr.random.next
        curr = curr.next.next

    # Step 3: Separate the copied list from the original
    curr = head
    copied_head = head.next
    while curr:
        copy = curr.next
        curr.next = copy.next
        copy.next = copy.next.next if copy.next else None
        curr = curr.next

    return copied_head
```

```python
node1 = RandomListNode(7)
node2 = RandomListNode(13)
node3 = RandomListNode(11)
node1.next = node2
node2.next = node3
node2.random = node1
node3.random = node1

copied_head = copyRandomList(node1)

# Print copied list values and random pointers
curr = copied_head
while curr:
    random_val = curr.random.val if curr.random else None
    print(f"Val: {curr.val}, Random: {random_val}")
    curr = curr.next
```

# **Valid Parentheses**

Input: "()[]{}"

Output: True

Input: "(]"

Output: False


```python
def isValid(s):
    stack = []
    mapping = {')':'(', '}':'{', ']':'['}

    for char in s:
        if char in mapping.values():
            stack.append(char)
        elif char in mapping:
            if not stack or stack.pop() != mapping[char]:
                return False
        else:
            return False
    return not stack
```

# **Implement Stack Using List / Deque**

```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, val):
        self.stack.append(val)

    def pop(self):
        return self.stack.pop() if self.stack else None

    def top(self):
        return self.stack[-1] if self.stack else None

    def isEmpty(self):
        return len(self.stack) == 0
```

```python
s = Stack()
s.push(10)
s.push(20)
print(s.top())    # 20
print(s.pop())    # 20
print(s.isEmpty()) # False
```

# **Min Stack**

Input: push(2), push(0), push(3), push(0)

getMin() -> 0

pop()

getMin() -> 0

pop()

getMin() -> 0

pop()

getMin() -> 2

```python
class MinStack:
    def __init__(self):
        self.stack = []
        self.min_stack = []

    def push(self, val):
        self.stack.append(val)
        if not self.min_stack or val <= self.min_stack[-1]:
            self.min_stack.append(val)

    def pop(self):
        if self.stack:
            val = self.stack.pop()
            if val == self.min_stack[-1]:
                self.min_stack.pop()

    def top(self):
        return self.stack[-1] if self.stack else None

    def getMin(self):
        return self.min_stack[-1] if self.min_stack else None
```

```python
minStack = MinStack()
minStack.push(2)
minStack.push(0)
minStack.push(3)
minStack.push(0)
print(minStack.getMin())  # 0
minStack.pop()
print(minStack.getMin())  # 0
minStack.pop()
print(minStack.getMin())  # 0
minStack.pop()
print(minStack.getMin())  # 2
```

**Evaluate Reverse Polish Notation (RPN)**
Input: ["2","1","+","3","*"]

Output: 9

Explanation: (2 + 1) * 3 = 9


```python
def evalRPN(tokens):
    stack = []
    for token in tokens:
        if token in "+-*/":
            b = stack.pop()
            a = stack.pop()
            if token == '+':
                stack.append(a + b)
            elif token == '-':
                stack.append(a - b)
            elif token == '*':
                stack.append(a * b)
            else:  # Division
                stack.append(int(a / b))  # truncate toward zero
        else:
            stack.append(int(token))
    return stack[0]
```

```python
tokens = ["2","1","+","3","*"]
print(evalRPN(tokens))  # 9

tokens = ["4","13","5","/","+"]
print(evalRPN(tokens))  # 6
```

# **Sliding Window Maximum**

Input: nums = [1,3,-1,-3,5,3,6,7], k = 3

Output: [3,3,5,5,6,7]


```python
from collections import deque

def maxSlidingWindow(nums, k):
    dq = deque()
    result = []

    for i, n in enumerate(nums):
        while dq and dq[0] <= i - k:
            dq.popleft()
        while dq and nums[dq[-1]] < n:
            dq.pop()
        dq.append(i)
        if i >= k - 1:
            result.append(nums[dq[0]])
    return result
```

```python
nums = [1,3,-1,-3,5,3,6,7]
k = 3
print(maxSlidingWindow(nums, k))  # [3,3,5,5,6,7]
```

# **Implement Queue using Stacks**

```python
class MyQueue:
    def __init__(self):
        self.in_stack = []
        self.out_stack = []

    def push(self, x):
        self.in_stack.append(x)

    def pop(self):
        self.move()
        return self.out_stack.pop()

    def peek(self):
        self.move()
        return self.out_stack[-1]

    def empty(self):
        return not self.in_stack and not self.out_stack

    def move(self):
        if not self.out_stack:
            while self.in_stack:
                self.out_stack.append(self.in_stack.pop())
```

```python
queue = MyQueue()
queue.push(1)
queue.push(2)
print(queue.peek())  # 1
print(queue.pop())   # 1
print(queue.empty()) # False
```
