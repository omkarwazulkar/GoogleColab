# **Power Set**

```python
def subsets(set):
  result = []
  total_num = len(set)

  def backtract(i, path):
    if i == total_num:
      result.append(path[:])
      return

    path.append(set[i])
    backtract(i+1, path)
    path.pop()

    backtract(i+1, path)

  backtract(0, [])
  return result
```

```python
set_ = [1, 2, 3, 4, 5, 7]
subsets(set_)
```

# **Set Starting With Specific Element**

```python
def specific_subsets(set):
  result = []
  total_num = len(set)

  def backtract(i, path):
    if i == total_num:
      if path and path[0] == 1:
        result.append(path[:])
      return

    path.append(set[i])
    backtract(i+1, path)
    path.pop()

    backtract(i+1, path)

  backtract(0, [])
  return result
```

```python
set_ = [1, 2, 3, 4, 5, 7]
specific_subsets(set_)
```

# **Set With Specific Length**

```python
def specific_length_subsets(set, length):
  result = []
  total_num = len(set)

  def backtract(i, path):
    if i == total_num:
      if len(path) == length:
        result.append(path[:])
      return

    path.append(set[i])
    backtract(i+1, path)
    path.pop()

    backtract(i+1, path)

  backtract(0, [])
  return result
```

```python
set_ = [1, 2, 3, 4, 5, 7]
specific_length_subsets(set_, 2)
```

# **Set With Target**

```python
def target_subsets(set, target):
  result = []
  total_num = len(set)

  def backtract(i, sum, path):
    if i == total_num:
      if sum == target:
        result.append(path[:])
      return

    path.append(set[i])
    backtract(i+1, sum + set[i], path)
    path.pop()

    backtract(i+1, sum, path)

  backtract(0, 0, [])
  return result
```

```python
set_ = [1, 2, 3, 4, 5, 7]
target_subsets(set_, 10)
```

# **Binary String With Specific Lenght & Start With**

```python
def binary_strings_start_with_1(n):
    result = []

    def backtrack(path):
        if len(path) == n:
            result.append("".join(path))
            return

        path.append('0')
        backtrack(path)
        path.pop()

        path.append('1')
        backtrack(path)
        path.pop()

    backtrack(['1'])
    return result
```

```python
binary_strings_start_with_1(3)
```
