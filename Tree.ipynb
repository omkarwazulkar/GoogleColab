# **Default Tree**

```
class TreeNode:
  def __init__(self, val):
    self.val = val
    self.left = None
    self.right = None
```

```
# Level 0
root = TreeNode(1)

# Level 1
root.left = TreeNode(2)
root.right = TreeNode(3)

# Level 2
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

# Level 3
root.left.left.left = TreeNode(8)
root.left.left.right = TreeNode(9)
root.left.right.left = TreeNode(10)
root.left.right.right = TreeNode(11)
root.right.left.left = TreeNode(12)
root.right.left.right = TreeNode(13)
root.right.right.left = TreeNode(14)
root.right.right.right = TreeNode(15)
```

# **BFS Level Order**

```
from collections import deque

def bfs_level_order(root):
  if not root:
    return []

  queue = deque([root])
  result = []

  while queue:

    level_size = len(queue)
    level_nodes = []

    for _ in range(level_size):

      node = queue.popleft()
      level_nodes.append(node.val)

      if node.left:
        queue.append(node.left)
      if node.right:
        queue.append(node.right)

    result.append(level_nodes)
  return result
```

```
bfs_level_order(root)
```

# **Sum/Minus**

```
from collections import deque

def bfs_level_order_sum(root):
  if not root:
    return []

  queue = deque([root])
  result = []

  while queue:

    level_size = len(queue)
    level_sum = 0

    for _ in range(level_size):

      node = queue.popleft()
      level_sum = level_sum + (node.val)

      if node.left:
        queue.append(node.left)
      if node.right:
        queue.append(node.right)

    result.append(level_sum)
  return result
```

```
bfs_level_order_sum(root)
```

# **Min/Max**

```
from collections import deque

def bfs_level_order_max(root):
  if not root:
    return []

  queue = deque([root])
  result = []

  while queue:

    level_size = len(queue)
    level_max = float('-inf')

    for _ in range(level_size):

      node = queue.popleft()
      level_max = max(level_max, node.val)

      if node.left:
        queue.append(node.left)
      if node.right:
        queue.append(node.right)

    result.append(level_max)
  return result
```

```
bfs_level_order_max(root)
```

# **Zig-Zag**

```
from collections import deque

def bfs_level_order_zig_zag(root):
  if not root:
    return []

  queue = deque([root])
  result = []
  left_to_right = True

  while queue:

    level_size = len(queue)
    level_nodes = []

    for _ in range(level_size):

      node = queue.popleft()
      level_nodes.append(node.val)

      if node.left:
        queue.append(node.left)
      if node.right:
        queue.append(node.right)

    if not left_to_right:
      level_nodes.reverse()

    result.append(level_nodes)
    left_to_right = not left_to_right

  return result
```

```
bfs_level_order_zig_zag(root)
```

# **Reverse Level Order Traversal / Reverse BFS Travesal / Bootom-Up Level Order Traversal**

```
from collections import deque

def bfs_level_order_reversal(root):
  if not root:
    return []

  queue = deque([root])
  result = []

  while queue:

    level_size = len(queue)
    level_nodes = []

    for _ in range(level_size):

      node = queue.popleft()
      level_nodes.append(node.val)

      if node.left:
        queue.append(node.left)
      if node.right:
        queue.append(node.right)

    result.append(level_nodes)
  return result[::-1]
```

```
bfs_level_order_reversal(root)
```

# **Maximum Depth**

```
from collections import deque

def max_depth_bfs(root):
    if not root:
        return 0

    queue = deque([root])
    depth = 0

    while queue:
        level_size = len(queue)
        for _ in range(level_size):
            node = queue.popleft()
            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)
        depth += 1  # finished processing one level

    return depth
```

```
def max_depth_dfs(root):
    if not root:
        return 0
    left_depth = max_depth_dfs(root.left)
    right_depth = max_depth_dfs(root.right)
    return 1 + max(left_depth, right_depth)
```

```
max_depth_bfs(root)
```

# **Check Symmetry**

```
from collections import deque

def is_symmetric_bfs(root):
    if not root:
        return True

    queue = deque([(root.left, root.right)])

    while queue:
        t1, t2 = queue.popleft()
        if not t1 and not t2:
            continue
        if not t1 or not t2:
            return False
        if t1.val != t2.val:
            return False
        queue.append((t1.left, t2.right))
        queue.append((t1.right, t2.left))

    return True
```

```
is_symmetric_bfs(root)
```

# **Invert Binary Trre**

```
from collections import deque

# BFS Invert
def invert_tree_bfs(root):
    if not root:
        return None

    queue = deque([root])

    while queue:
        node = queue.popleft()
        # Swap left and right
        node.left, node.right = node.right, node.left

        if node.left:
            queue.append(node.left)
        if node.right:
            queue.append(node.right)

    return root

# Level-Order
def print_tree_level_order(root):
    if not root:
        print("Empty tree")
        return

    queue = deque([root])
    while queue:
        level_size = len(queue)
        level_nodes = []
        for _ in range(level_size):
            node = queue.popleft()
            level_nodes.append(node.val if node else None)
            if node:
                queue.append(node.left)
                queue.append(node.right)
        print(level_nodes)


print("Original Tree:")
print_tree_level_order(root)

# Invert tree
invert_tree_bfs(root)

print("\nInverted Tree:")
print_tree_level_order(root)
```

# **Subset Target**

```
def target_subset(nums, target):

  result = []
  n = len(nums)

  def backtrack(i, sum, path):
    if i == n:
      if sum == target:
        result.append(path[:])
      return

    path.append(nums[i])
    backtrack(i + 1, sum + nums[i], path)
    path.pop()

    backtrack(i+1, sum, path)

  backtrack(0, 0, [])
  return result
```

```
set_ = [1, 2, 5, 3, 4, 6, 8]
target_subset(set_, 10)
```
