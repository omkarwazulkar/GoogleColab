```python
class Node:
  def __init__(self, data):
    self.data = data
    self.next = None
```

```python
class LinkedList:
  def __init__(self):
    self.head = None

  def insert_begin(self, data):
    new_node = Node(data)
    new_node.next = self.head
    self.head = new_node

  def insert_end(self, data):
    if self.head is None:
      self.head = Node(data)
      return

    temp = self.head
    while temp.next is not None:
      temp = temp.next

    temp.next = Node(data)

  def delete_begin(self):
    if self.head is None:
        print("List is empty")
        return

    self.head = self.head.next

  def delete_end(self):
    if self.head is None:
      return ("List Empty")

    if self.head.next is None:
      return

    temp = self.head
    while temp.next.next is not None:
      temp = temp.next
    temp.next = None

  def delete_key(self, data):

    if self.head is None:
      return ("List Empty")

    if self.head.data == data:
      return

    temp = self.head
    while temp.next is not None:
      if temp.next.data == data:
        temp.next = temp.next.next
        return
      temp = temp.next
    print("Key Not Found")

  def count_length(self):

    count = 0

    if self.head is None:
      return 0

    temp = self.head
    while temp.next is not None:
      count += 1
      temp = temp.next
    return count+1

  def search_key(self, key):
    if self.head is None:
      return "No Key"

    temp = self.head
    while temp.next is not None:
      if temp.data == key:
        return True
      temp = temp.next
    return "Key Not Found"

  def print_(self):
    temp = self.head
    while temp:
      print(temp.data, end = " -> ")
      temp = temp.next
    print(None)

  def reverse_iterative(self):
    prev = None
    curr = self.head

    while curr:
        nxt = curr.next
        curr.next = prev
        prev = curr
        curr = nxt

    self.head = prev
```

```python
def merge_sorted(l1, l2):
  dummy = Node(0)
  tail = dummy
  while l1 and l2:
    if l1.data <= l2.data:
      tail.next = l1
      l1 = l1.next
    else:
      tail.next = l2
      l2 = l2.next

    tail = tail.next

  if l1:
      tail.next = l1
  else:
      tail.next = l2

  return dummy.next
```

```python
def has_cycle(self):
    slow = fast = self.head

    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

        if slow == fast:
            return True

    return False
```

```python
def remove_cycle(self):
    slow = fast = self.head

    # Step 1: detect cycle
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
        if slow == fast:
            break
    else:
        return  # no cycle present

    # Step 2: find start of cycle
    slow = self.head
    while slow != fast:
        slow = slow.next
        fast = fast.next

    # Step 3: find last node inside loop
    while fast.next != slow:
        fast = fast.next

    fast.next = None  # break loop
```

```python
def reverse_iterative(self):
    prev = None
    curr = self.head

    while curr:
        nxt = curr.next
        curr.next = prev
        prev = curr
        curr = nxt

    self.head = prev
```

```python
def reverse_recursive(self, node):
    if node is None or node.next is None:
        return node

    rest = self.reverse_recursive(node.next)

    node.next.next = node
    node.next = None                  # break old link

    return rest
```

```python
def nth_from_end(self, n):
    slow = fast = self.head

    # move fast n steps
    for _ in range(n):
        if not fast:
            return "n is greater than list length"
        fast = fast.next

    # move both step-by-step
    while fast:
        slow = slow.next
        fast = fast.next

    return slow.data   # slow is at nth from end
```

```python
def is_palindrome(self):
    if not self.head or not self.head.next:
        return True

    # Step 1: find middle
    slow = fast = self.head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next

    # Step 2: reverse second half
    prev = None
    curr = slow
    while curr:
        nxt = curr.next
        curr.next = prev
        prev = curr
        curr = nxt

    # Step 3: compare halves
    left = self.head
    right = prev   # head of reversed part

    while right:
        if left.data != right.data:
            return False
        left = left.next
        right = right.next

    return True
```

```python
def get_intersection(head1, head2):
    # count lengths
    def length(head):
        c = 0
        while head:
            c += 1
            head = head.next
        return c

    len1 = length(head1)
    len2 = length(head2)

    # move longer list ahead
    d = abs(len1 - len2)

    if len1 > len2:
        for _ in range(d): head1 = head1.next
    else:
        for _ in range(d): head2 = head2.next

    # move both until intersection
    while head1 and head2:
        if head1 == head2:
            return head1.data   # intersection point
        head1 = head1.next
        head2 = head2.next

    return None
```
