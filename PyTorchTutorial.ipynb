```
import torch
import torch.nn as nn
import torch.optim as optim

# Example data: 2 images, each 4x3
x_train = torch.tensor([[[2.0, 6.0, 8.0],
                         [3.0, 9.0, 12.0],
                         [4.0, 12.0, 16.0],
                         [5.0, 15.0, 20.0]],

                        [[2.0, 6.0, 8.0],
                         [3.0, 9.0, 12.0],
                         [4.0, 12.0, 16.0],
                         [5.0, 15.0, 20.0]]], dtype=torch.float32)

y_train = torch.tensor([[4.0], [6.0]], dtype=torch.float32)

# Flatten input: each image  12 features
x_train_flat = x_train.view(2, -1) / 20.0  # normalize

# Design neural network using Sequential
model = nn.Sequential(
    nn.Linear(12, 8),
    nn.ReLU(),
    nn.Linear(8, 4),
    nn.ReLU(),
    nn.Linear(4, 1)
)

# Loss and optimizer
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Training loop
for step in range(20):
    y_pred = model(x_train_flat)
    loss = criterion(y_pred, y_train / 10.0)  # normalize target
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    print(f"Step {step+1}: Loss = {loss.item():.6f}")
```

```
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```
y = df["MEDV"]
```

```
import torch

X_t = torch.tensor(X_scaled.values, dtype=torch.float32)
y_t = torch.tensor(y.values, dtype=torch.float32).view(-1, 1)
```

```
X_t.shape, y_t.shape
```

```
y_t[0]
```

```
import torch.nn as nn

class BostonNN(nn.Module):
  def __init__(self):
    super(BostonNN, self).__init__()
    self.model = nn.Sequential(
        nn.Linear(13, 32),
        nn.ReLU(),
        nn.Linear(32, 1)
    )

  def forward(self, x):
    return self.model(x)
```

```
net = BostonNN()
```

```
net
```

```
import torch.optim as optim

criterion = nn.MSELoss()
optimizer = optim.Adam(net.parameters(), lr=0.1)
```

```
num_epochs = 1000

for epoch in range(num_epochs):
  net.train()

  y_pred = net(X_t)
  loss = criterion(y_pred, y_t)
  optimizer.zero_grad()
  loss.backward()
  optimizer.step()

  if epoch%20==0:
    print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
```

```
net.eval()
with torch.no_grad():
    val_pred = net(X_t)
    print("Sample predictions:", val_pred[:5].squeeze())
```
